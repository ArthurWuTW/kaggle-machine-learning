```python
def get_thresholds(index, dataset):
    # sorted data by index-th column
    column = dataset.columns[index]
    sorted_dataset = dataset.sort_values(by=[column], ascending=True)
    print("Column", column)
    values = sorted_dataset[column].drop_duplicates().to_numpy()
    print("values", values)
    samples = []
    for i in range(len(values)+1):
        if(i==0):
            samples.append(values[i]-1)
        elif(i==len(values)):
            samples.append(values[i-1]+1)
        else:
            samples.append((values[i]+values[i-1])/2)
    print("samples",samples)
    for sample in samples:
        print("-----------------------")
        print("sample=", sample)
        left = dataset[dataset[column] < sample]
        right = dataset[dataset[column] > sample]
        print("left num=", len(left));print(left);print("right, num=", len(right));print(right);
    return sorted_dataset
```
