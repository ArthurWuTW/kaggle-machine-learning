```python
def get_thresholds(index, dataset):
    # sorted data by index-th column
    column = dataset.columns[index]
    sorted_dataset = dataset.sort_values(by=[column], ascending=True)
    print("Column", column)
    values = sorted_dataset[column].drop_duplicates().to_numpy()
    
    gt_column = dataset.columns[len(dataset.columns)-1]
    gt_values = sorted_dataset[gt_column].drop_duplicates().to_numpy()
    
    samples = []
    for i in range(len(values)+1):
        if(i==0):
            samples.append(values[i]-1)
        elif(i==len(values)):
            samples.append(values[i-1]+1)
        else:
            samples.append((values[i]+values[i-1])/2)
    print("samples",samples)
    for sample in samples:
        print("-----------------------")
        print("sample=", sample)
        left = dataset[dataset[column] < sample]
        right = dataset[dataset[column] > sample]
        # print("left num=", len(left));print(left);print("right, num=", len(right));print(right);
        get_gini_index(left, right, gt_values, gt_column)
    return sorted_dataset

def get_gini_index(left_leaf, right_leaf, result_values, result_column_name):
    left_gini = 0.0
    if(len(left_leaf)!=0):
        left_gini = 1.0
        for result in result_values:
            pi_N = len(left_leaf[left_leaf[result_column_name] == result])/len(left_leaf)
            left_gini -= (pi_N) **2
            print("left result", result, "pi_N", pi_N, "size", len(left_leaf[left_leaf[result_column_name] == result]), "total", len(left_leaf))
    right_gini = 0.0
    if(len(right_leaf)!=0):
        right_gini = 1.0
        for result in result_values:
            pi_N = len(right_leaf[right_leaf[result_column_name] == result])/len(right_leaf)
            right_gini -= (pi_N) **2
            print("right", "result", result, "pi_N", pi_N, "size", len(right_leaf[right_leaf[result_column_name] == result]), "total", len(right_leaf))

    total = len(left_leaf) + len(right_leaf)
    weighted_gini = left_gini*len(left_leaf)/total + right_gini*len(right_leaf)/total
    print("left_gini", left_gini, "right_gini", right_gini, "weighted_gini", weighted_gini)
    return weighted_gini
```
